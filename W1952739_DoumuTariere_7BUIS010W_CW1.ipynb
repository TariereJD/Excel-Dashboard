```python
# connect to database
conn = sqlite3.connect("COVID_19")
cur = conn.cursor()
```

```python
# load dataframes into database As tables with names Fact,DT,SO,VA, LO
df_Fact.to_sql("Fact", conn, if_exists="replace")
df_DT.to_sql('DT',conn, if_exists="replace")
df_SO.to_sql('SO',conn, if_exists="replace")
df_LO.to_sql('LO',conn, if_exists="replace")
df_VA.to_sql('VA',conn, if_exists="replace")
```

```python
# CHECK IF DATA INSERTED INTO TABLES Fact,DT,SO,VA, LO
Fact = pd.read_sql('SELECT * FROM Fact', conn)
Fact
```

```python
#CREATING A DATA MART THAT CAN GENERATE REPORTS OF DIFFERENT QUESTIONS
M1 = pd.read_sql('''SELECT Fact.L_ID,Fact.T_ID,SO.Source_ID,VA.Vacc_ID,SUM(Current_cases) + SUM(New_cases) as Total_cases, 
SUM(Deaths) as Total_deaths, SUM(Doses_recieved) as doses_received ,SUM(ICU_cases) as ICU_cases,LO.Region, LO.Country, DT.Month, DT.Quater,
(SUM(Doses_recieved)/LO.Population) as vaccination_rate FROM Fact
INNER JOIN  LO ON LO.L_ID = Fact.L_ID 
INNER JOIN DT ON DT.T_ID = Fact.T_ID
INNER JOIN SO ON SO.Source_ID = Fact.Source_ID
INNER JOIN VA ON VA.Vacc_ID = Fact.Vacc_ID
GROUP BY LO.Region, LO.Country, DT.Month,DT.Quater
''', conn)
M1
```

```python
#Creaating the materialized view
cur.execute('''CREATE VIEW M1 AS
SELECT Fact.L_ID,Fact.T_ID,SO.Source_ID,VA.Vacc_ID,SUM(Current_cases) + SUM(New_cases) as Total_cases, SUM(Deaths) as Total_deaths,
 SUM(Doses_recieved) as doses_received ,SUM(ICU_cases) as ICU_cases,LO.Region, LO.Country, DT.Month, DT.Quater,
(SUM(Doses_recieved)/LO.Population) as vaccination_rate FROM Fact
INNER JOIN  LO ON LO.L_ID = Fact.L_ID 
INNER JOIN DT ON DT.T_ID = Fact.T_ID
INNER JOIN SO ON SO.Source_ID = Fact.Source_ID
INNER JOIN VA ON VA.Vacc_ID = Fact.Vacc_ID
GROUP BY LO.Region, LO.Country, DT.Month,DT.Quater
''')
```

```python
#ANSWER TO QUESTION 1, CREATING A TABLE THAT IS GROUP BY REGION AND MONTH AND ALSO SHOWS THE VACCINE RATE OF EACH VACCINE 
Q1 = pd.read_sql('''SELECT Total_cases, Total_deaths ,ICU_cases,vaccination_rate, Region, Month,Vacc_ID,Source_ID
FROM M1
GROUP BY Region, Month
''',conn)

Q1
```

```python
#CREATING A TABLE THAT SHOWS JUST GROUP BY REGION TO VISUALIZE THE REGIONS AND THE COVID DESCRIPTIONS
Q1a = pd.read_sql('''SELECT Total_cases, Total_deaths ,ICU_cases,vaccination_rate, Region, Month,Vacc_ID
FROM M1
GROUP BY Region
''',conn)
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
df = pd.DataFrame({
    'Total cases':Q1a['Total_cases'],
    'Total deaths': Q1a['Total_deaths'],
    'ICU cases': Q1a['ICU_cases'],
    'Region': Q1a['Region'],
})
  
# plotting graph
df.plot(y=["Total cases","Total deaths","ICU cases"], x="Region", kind="bar",figsize=(16,8))
```
